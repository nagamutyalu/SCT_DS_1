```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_559803.csv")
data.head()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Region", data=data, order=data["Region"].value_counts().index, palette="viridis")
plt.title("Number of Countries by Region")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x="IncomeGroup", data=data, order=data["IncomeGroup"].value_counts().index, palette="Set2")
plt.title("Number of Countries by Income Group")
plt.show()
```
